```python
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_data = pd.read_csv('datasets/train.csv')
test_data = pd.read_csv('datasets/test.csv')
```

```python
# Шаг 1. добавил разграниение для тестовых и тренировочных данных
train_data['is_test'] = 0
test_data['is_test'] = 1
```

```python
# Шаг 2. объеденим данные
titanic_data = train_data.append(test_data, ignore_index=True)
titanic_data
```

```python
# Шаг 3. Проанализируем данные  
# проверяем пустые значения
titanic_data.isnull().sum()
```

```python
titanic_data
```

```python
# Делаем выводы
# 1. Cabin имеет 77% пустых значений - удаляем
# 2. Name и Ticket не несет полезной информации - удаляем
# 3. SibSp и Parch стоит объеденить, так как несут похожий смысл - количество родственников на корабле
# 4. Sex и Embarked - нужно привести к числу
# 5. В Age 25% пустых значений - нужно восстановить
```

```python
# Шаг 4. Удаляем ненужные поля
titanic_data = titanic_data.drop(['Name', 'Ticket', 'Cabin'], axis=1)
titanic_data
```

```python
# Шаг 5. Считаем количество родственников и удаляем лишнее
titanic_data['family_size'] = titanic_data["Parch"] + titanic_data["SibSp"]
titanic_data = titanic_data.drop(["Parch", "SibSp"], axis=1)
titanic_data
```

```python
# Шаг 6. Приводим все к числовым значениям
titanic_data = pd.get_dummies(titanic_data)
titanic_data = titanic_data.drop(['Sex_female'], axis=1)
titanic_data
```

```python
# Шаг 7. Восстанавливаем возраст
# Посчитаем средний возраст для ключа (класс, пол)
age_info = titanic_data.groupby(['Sex_male','Pclass'])
age_info.Age.median()
```

```python
titanic_data.Age = age_info.Age.apply(lambda x: x.fillna(x.median()))
titanic_data
```

```python
# Шаг 8. Заменим Fare на средний
titanic_data["Fare"] = titanic_data["Fare"].fillna(titanic_data["Fare"].mean())
```

```python
# данные готовы для посторения модели
titanic_data.isnull().sum()
```

```python
train_data = titanic_data.query("is_test == 0")
test_data = titanic_data.query("is_test == 1")
```

```python
# Шаг 9. Выбросим порт отправления
# Я не увидел причин оставить порт отправления в данных: влияние скорее рандомное
titanic_data = titanic_data.drop(['Embarked_S', "Embarked_Q", "Embarked_C"], axis=1)

train_data = titanic_data.query("is_test == 0")
test_data = titanic_data.query("is_test == 1")

train_data = train_data.drop(['is_test'], axis=1)
test_data = test_data.drop(['is_test'], axis=1)
```

```python
train_data
```

```python
test_data
```

```python
# Шаг 10. Готовим данные

# для обучения
x = train_data.drop('Survived', axis=1).values 
y = train_data["Survived"].values
y = y.astype("int")

# для теста
x_test = test_data.drop('Survived', axis=1).values
```

```python
# Шаг 11. Переберем параметры: 
# максимальную глубинку дерева;
# количесто деревьев;
# минимальное количество объектов, необходимое для разделения внутреннего узла;
# минимальное число объектов в листе. 

forrest_params = dict(     
    max_depth = [n for n in range(5, 20)],
    n_estimators = [n for n in range(5, 20)],
    min_samples_split = [n for n in range(3, 12)], 
    min_samples_leaf = [n for n in range(4, 8)],   
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Шаг 12. Создаем случайного леса 
forrest_clf = RandomForestClassifier()
```

```python
# Шаг 13. Обучаем модель
forest_cv = GridSearchCV(estimator=forrest_clf, param_grid=forrest_params, cv=5) 
forest_cv.fit(x, y)
```

```python
print(f"Результат: {forest_cv.best_score_}")
print(f"Параметры: {forest_cv.best_estimator_}")
```

```python
forrest_pred = forest_cv.predict(x_test)
```

```python
# Шаг 14. Вывод результаа
passengerId = test_data["PassengerId"]
result = pd.DataFrame({'PassengerId': passengerId, 'Survived': forrest_pred})
result.to_csv('titanic.csv', index=False)
```
