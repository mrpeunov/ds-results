```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('digits_datasets/train.csv')
test_data = pd.read_csv('digits_datasets/test.csv')
```

```python
# Шаг 1. Проверим пропуски
sum(train_data.isnull().sum())    
```

```python
sum(test_data.isnull().sum())   
```

```python
# Шаг 2. Подготовим данные
x = train_data.drop('label', axis=1).values
y = train_data['label'].values
```

```python
# Шаг 3. Переберем максимальную глубину и количество деревьев
forrest_params = dict(     
    max_depth = [n for n in range(10, 70, 10)],
    n_estimators = [n for n in range(140, 160, 5)]
)
```

```python
# Шаг 4. Создаем случайный лес
forrest_clf = RandomForestClassifier()
```

```python
# Шаг 5. Обучаем модель
forest_cv = GridSearchCV(estimator=forrest_clf, param_grid=forrest_params, cv=5) 
forest_cv.fit(x, y)
```

```python
print(f"Результат: {forest_cv.best_score_}")
print(f"Параметры: {forest_cv.best_estimator_}")
```

```python
# Шаг 6. Считаем результат на тестовых данных
forrest_pred = forest_cv.predict(test_data)
```

```python
# Шаг 7. Выводим результат
result = pd.DataFrame(
    {
        "ImageId": range(1,len(forrest_pred)+1),
        "Label": forrest_pred
    }
)
result.to_csv('digits.csv', index=False, header=True)
```
